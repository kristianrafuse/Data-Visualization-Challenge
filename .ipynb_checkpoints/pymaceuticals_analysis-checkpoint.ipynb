```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_df = df = pd.merge(study_results, mouse_metadata, on=["Mouse ID"])

# Display the data table for preview
merged_df.sample(25)
```

```python
# Checking the number of mice.
merged_df['Mouse ID'].nunique()
# or len(merged_df['Mouse ID'].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
grouped_df = merged_df.groupby(['Mouse ID', 'Timepoint'])
counts = grouped_df.size()
grouped_df.size()
```

```python
# Find any mouse ID with duplicate time points
duplicates = counts[counts > 1]
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID.
g989_data = merged_df.loc[(merged_df["Mouse ID"] == 'g989'), :]
g989_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# Checking the number of mice in the clean DataFrame.
clean_df = merged_df.loc[(merged_df["Mouse ID"] != 'g989'), :]
print(clean_df['Mouse ID'].nunique())
clean_df
```

## Summary Statistics

```python
mean_tumor = merged_df.groupby(["Drug Regimen"]).mean()['Tumor Volume (mm3)']
meidan_tumor = merged_df.groupby(["Drug Regimen"]).median()['Tumor Volume (mm3)']
var_tumor = meidan_tumor = merged_df.groupby(["Drug Regimen"]).var()['Tumor Volume (mm3)']
stddev_tumor = meidan_tumor = merged_df.groupby(["Drug Regimen"]).std()['Tumor Volume (mm3)']
stderr_tumor = meidan_tumor = merged_df.groupby(["Drug Regimen"]).sem()['Tumor Volume (mm3)']

summarydata = {
    'Mean Tumor Volume':mean_tumor,
    'Median Tumor Volume': meidan_tumor,
    'Tumor Volume Variance':var_tumor,
    'Tumor Volume Std. Dev.':stddev_tumor,
    'Tumor Volume Std. Err.':stderr_tumor}

summary_df = pd.DataFrame(summarydata)
summary_df
```

```python
treatments = merged_volumes_df['Drug Regimen'].unique().tolist()

tumor_vol_data = {}
for treatment in treatments:
    tumor_vol_data[treatment] = []
    
grouped_data = merged_volumes_df.groupby(['Drug Regimen', 'Mouse ID'])
for (treatment, mouse), treatment_data in grouped_data:
    if treatment in treatments:
        tumor_vol_data[treatment].append(treatment_data.iloc[-1]['Tumor Volume (mm3) Final'])

tumor_vol_data
```

```python
#Calculate the IQRs per Drug Regimen
iqr_df = pd.DataFrame.from_dict(tumor_vol_data)

q1 = iqr_df.quantile(0.25)
q3 = iqr_df.quantile(0.75)

iqr = q3 - q1

iqr
```

```python
# Determine upper and lower bounds
lowerbound = q1 - 1.5*iqr
upperbound = q3 + 1.5*iqr

print(lowerbound)
print(upperbound)
```

```python
# Determine outliers from upper and lower bounds
outliers = {}
for col in iqr_df.columns:
    outliers[col] = iqr_df[(iqr_df[col] > upperbound[col]) 
    | (iqr_df[col] < lowerbound[col])][col].tolist()
    
outliers
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
outlier_props = dict(marker='o', markerfacecolor='red')
ax = iqr_df.boxplot(figsize=(8,6), flierprops=outlier_props, grid=False)
ax.set_xlabel('Treatment Group')
ax.set_ylabel('Tumor Volume (mm3)')
ax.set_title('Distribution of Tumor Volume for Each Treatment Group')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

slope, intercept, r_value, p_value, std_err = st.linregress(avg_tumor_vol_df['Weight (g)'], avg_tumor_vol_df['Tumor Volume (mm3)'])
print(slope, intercept, r_value, p_value, std_err)
regress_values = avg_tumor_vol_df['Weight (g)'] * slope + intercept
print(regress_values)
line_eq = "y = " + str(slope) + "x +" + str(intercept)
print(line_eq)
```
